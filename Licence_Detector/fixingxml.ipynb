```python
def fixingxml_ownsamples(datasource,n):
    tree =ET.parse(datasource)
    root = tree.getroot()
    b2 = root.find("./object")
    # creating a new element/xml-node
    c = ET.Element("occluded")
    c.text = "0"
    index = list(b2).index(list(b2)[3])#index==3

    b2.insert(index, c)
    root.remove(root.find("./source"))
    root.remove(root.find("./path"))
    tree2 = ET.ElementTree(root)
    tree2.write("D:\Anaconda_en_DiscoD\car_plate_detection2\TFODCourse\Tensorflow\workspace\images\\xml_and_images\\"+str(n))
```
